```python
class Graph:
    def __init__(self, nVertices):
        self.nVertices = nVertices
        self.adjMatrix = [[0 for i in range(nVertices)] for j in range(nVertices)]
           
    def addEdge(self, v1, v2):
        self.adjMatrix[v1][v2] = 1
        self.adjMatrix[v2][v1] = 1
    
    def getPath(self, v1, v2):        
        parent = {}
        visited = [False] * self.nVertices
        
        queue = []
        queue.append(v1)
        visited[v1] = True
        
        while queue:
            front = queue.pop(0)
                                    
            if front == v2:
                return parent
            
            for i in range(self.nVertices):
                if self.adjMatrix[front][i] > 0 and visited[i] is False:
                    queue.append(i)
                    visited[i] = True
                    parent[i] = front  
                    
        return None
    
    
v, e = [int (i) for i in input().split()[:2]]
g = Graph(v)

for i in range(e):
    a, b = [int(x) for x in input().split()[:2]]
    g.addEdge(a, b)    
v1, v2 = [int (r) for r in input().split()[:2]]

parent = g.getPath(v1, v2)
#print(parent)


listValue = [v2]
prev = parent[v2]
listValue.append(prev)
while (1):
    prev = parent[prev]
    listValue.append(prev)
    if prev == v1:
        break
    
for j in listValue:
    print(j, end = " ")
```
