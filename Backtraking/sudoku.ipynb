```python
def find_Zero(board):
    for i in range(9):
        for j in range(9):
            if board[i][j] == 0:
                return (i,j)
        
    return None


def validate(row,col,board,number):
    ##row check
    for column in range(9):
        if board[row][column] == number and column != col:
            return False
        
    #column check
    for rowCheck in range(9):
        if board[rowCheck][col] == number and rowCheck != row:
            return False
        
    box_x = row // 3
    box_y = col // 3
    for i in range(box_y*3,box_y*3 + 3):
        for j in range(box_x*3,box_x*3 + 3):
            if board[i][j] == number and row != i and j != col:
                return False
           
    print(board)
    return True
        


def validateSudoku(row,col,board):
    find = find_Zero(board)
    if find is None:
        print(board)
        return True
    else:
        row,col = find
        for i in range(1,10):
            isValid = validate(row,col,board,i)
            if isValid:
                board[row][col] = i
                if validateSudoku(row,col,board):
                    return True
                board[row][col] = 0
                
        return False
                


def solveSudoku(board):
    valid = validateSudoku(0,0,board)
    return valid

#board = [[ int(ele) for ele in input().split() ]for i in range(9)]
board = [[9, 0, 0, 0, 2, 0, 7, 5, 0], 
[6, 0, 0, 0, 5, 0, 0, 4, 0 ],
[0, 2, 0, 4, 0, 0, 0, 1, 0 ],
[2 ,0 ,8 ,0 ,0 ,0 ,0 ,0 ,0 ],
[0 ,7 ,0 ,5 ,0 ,9 ,0 ,6 ,0 ],
[0 ,0 ,0 ,0 ,0 ,0 ,4 ,0 ,1 ],
[0 ,1 ,0 ,0 ,0 ,5 ,0 ,8 ,0 ],
[0 ,9 ,0 ,0 ,7, 0 ,0 ,0 ,4 ],
[0 ,8 ,2 ,0 ,4 ,0 ,0 ,0 ,6]]
ans = solveSudoku(board)
if ans is True:
    print('true')
else:
    print('false')
```

```python
board = [
    [7,8,0,4,0,0,1,2,0],
    [6,0,0,0,7,5,0,0,9],
    [0,0,0,6,0,1,0,7,8],
    [0,0,7,0,4,0,2,6,0],
    [0,0,1,0,5,0,9,3,0],
    [9,0,4,0,6,0,0,0,5],
    [0,7,0,3,0,0,0,1,2],
    [1,2,0,0,0,7,4,0,0],
    [0,4,9,2,0,6,0,0,7]
]


def solve(bo):
    find = find_empty(bo)
    if not find:
        return True
    else:
        row, col = find

    for i in range(1,10):
        if valid(bo, i, (row, col)):
            bo[row][col] = i

            if solve(bo):
                return True

            bo[row][col] = 0

    return False


def valid(bo, num, pos):
    # Check row
    for i in range(len(bo[0])):
        if bo[pos[0]][i] == num and pos[1] != i:
            return False

    # Check column
    for i in range(len(bo)):
        if bo[i][pos[1]] == num and pos[0] != i:
            return False

    # Check box
    box_x = pos[1] // 3
    box_y = pos[0] // 3

    for i in range(box_y*3, box_y*3 + 3):
        for j in range(box_x * 3, box_x*3 + 3):
            if bo[i][j] == num and (i,j) != pos:
                return False

    return True


def find_empty(bo):
    for i in range(len(bo)):
        for j in range(len(bo[0])):
            if bo[i][j] == 0:
                return (i, j)  # row, col

    return None

board = [[9, 0, 0, 0, 2, 0, 7, 5, 0], 
[6, 0, 0, 0, 5, 0, 0, 4, 0 ],
[0, 2, 0, 4, 0, 0, 0, 1, 0 ],
[2 ,0 ,8 ,0 ,0 ,0 ,0 ,0 ,0 ],
[0 ,7 ,0 ,5 ,0 ,9 ,0 ,6 ,0 ],
[0 ,0 ,0 ,0 ,0 ,0 ,4 ,0 ,1 ],
[0 ,1 ,0 ,0 ,0 ,5 ,0 ,8 ,0 ],
[0 ,9 ,0 ,0 ,7, 0 ,0 ,0 ,4 ],
[0 ,8 ,2 ,0 ,4 ,0 ,0 ,0 ,6]]
ans = solve(board)
if ans is True:
    print('true')
else:
    print('false')
```
