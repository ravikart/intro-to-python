```python
#Power x ** n
def power(x, n):
    if n == 0:
        return 1
    ans = power(x,n//2)
    print(ans,n,sep=" ")
    if n % 2 == 0:
        return ans*ans
    else:
        return x * ans * ans

# Main
from sys import setrecursionlimit
setrecursionlimit(11000)
x, n=list(int(i) for i in input().strip().split(' '))
print(power(x, n))
```

```python
#Array itersection
def mergeList(listOne,listTwo,listValue):
    firstIterator = 0
    secondIterator = 0
    firstLength = len(listOne)
    secondLength = len(listTwo)
    mainIterator = 0
    while((firstIterator < firstLength) and (secondIterator < secondLength)):
        if(listOne[firstIterator] < listTwo[secondIterator]):
            listValue[mainIterator] = listOne[firstIterator]
            firstIterator = firstIterator + 1
            mainIterator = mainIterator + 1
        else:
            listValue[mainIterator] = listTwo[secondIterator]
            secondIterator = secondIterator + 1
            mainIterator = mainIterator + 1
    
    while(firstIterator < firstLength):
        listValue[mainIterator] = listOne[firstIterator]
        firstIterator = firstIterator + 1
        mainIterator = mainIterator + 1
        
    while(secondIterator < secondLength):
        listValue[mainIterator] = listTwo[secondIterator]
        secondIterator = secondIterator + 1
        mainIterator = mainIterator + 1
        

def mergeSort(listValue):
    length = len(listValue)
    mid = (length)//2
    if length == 0 or length == 1:
        return
    
    listOne = listValue[:mid]
    listTwo = listValue[mid:]
    
    mergeSort(listOne)
    mergeSort(listTwo)
    
    mergeList(listOne,listTwo,listValue)
    
    
    
def findIndex(element,listValue,start,end):
    if(start > end):
        return -1
    
    midValue = ((end - start)//2) + start
    if listValue[midValue] == element:
        return midValue
    elif listValue[midValue] < element:
        return findIndex(element,listValue,midValue+1,end)
    else:
        return findIndex(element,listValue,start,midValue-1)
    
    

def intersection(arr1, arr2):
    for element in arr1:
        index = findIndex(element,arr2,0,len(arr2)-1)
        if index != -1:
            print(element)
            del arr2[index]

# Main
n1=int(input())
arr1=list(int(i) for i in input().strip().split(' '))
n2=int(input())
arr2=list(int(i) for i in input().strip().split(' '))
if n1 > n2:
    mergeSort(arr1)
    intersection(arr2, arr1)
else:
    mergeSort(arr2)
    intersection(arr1, arr2)
```

```python
def calculateRightSum(listValue):
    sum = 0
    for element in listValue:
        sum = sum + element
    return sum


def equilibriumIndex(arr):
    iterator = 0
    length = len(arr)
    leftSum = 0
    rightSum = 0
    indexElement = arr[0]
    while(iterator < length):
        if(iterator == 0):
            rightSum = calculateRightSum(arr[1:])
            
        print(leftSum,rightSum,sep=" ")
        if leftSum == rightSum:
            return iterator
        else:
            leftSum = leftSum + indexElement
            if (iterator+1) < (length - 1):
                indexElement = arr[iterator + 1]
                print(indexElement)
                rightSum = rightSum - indexElement
        iterator = iterator + 1
    else:
        return -1
            

# Main
n = int(input())
arr = [int(i) for i in input().strip().split()]
print(equilibriumIndex(arr))
```

```python
def FindUnique(arr):
    emptyDict = {}
    for element in arr:
        value = emptyDict.get(element)
        if value == None :
            emptyDict[element] = 1
        else:
            emptyDict[element] = value + 1
    
    for mapValue in emptyDict:
        if(emptyDict.get(mapValue) == 1):
            return mapValue

# Main
n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
unique=FindUnique(arr)
print(unique)
```

```python
def Rotate(arr, d):
    for iterator in range(len(arr),d,-1):
        value = arr.pop()
        arr.insert(0,value)

# Main
n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
d=int(input())
Rotate(arr, d % n)
print(*arr)
```
